```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
```

```python
Matplotlib is a popular Python library used for data visualization. It provides a wide range of functionalities for creating various types of plots and charts, making it an essential tool for researchers, data scientists, and analysts.
1.Line Plot: A basic plot where data points are connected with straight lines, suitable for showing trends over time or continuous data.

2.Scatter Plot: Used to visualize the relationship between two variables, displaying individual data points as dots on the graph.

3.Bar Plot: Ideal for comparing categorical data, where bars represent the frequency or value of each category.

4.Histogram: Used to show the distribution of a dataset by dividing it into intervals (bins) and displaying the frequency of values within each bin.

5.Pie Chart: Suitable for illustrating the proportion of different categories in a dataset as slices of a pie, with each slice representing a category's percentage.
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
```

```python
A scatter plot is a type of data visualization used to display the relationship between two numerical variables. It is a graph in which individual data points are plotted as dots on a two-dimensional coordinate system, with one variable represented on the x-axis and the other variable represented on the y-axis.
```

```python
import numpy as np
import matplotlib.pyplot as  plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.scatter(x,y)
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It shows the summary statistics of the data, such as the median, quartiles, and potential outliers, in a compact and easy-to-understand manner. Box plots are particularly useful for identifying the spread and skewness of the data and for comparing multiple datasets side by side.
```

```python
plt.boxplot(box1)
plt.boxplot(box2)
```
